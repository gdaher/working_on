```python
#Source : https://medium.com/@msalmon00/web-scraping-job-postings-from-indeed-96bd588dcb4b

import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
URL = 'http://rss.indeed.com/rss?q=analyst&l=Boston%2C+MA'
# Get HTML results
page = requests.get(URL)

print(page.text)
```

```python
# Parse HTML into Beautiful Soup object
# Understanding parsers for beautiful soup:https://beautiful-soup-4.readthedocs.io/en/latest/
soup = BeautifulSoup(page.text)

# Print for easier reading
print(soup.prettify())
```

```python
def extract_job_title_from_result(soup): 
    jobs = []
    for item in soup.find_all(name="item"):
      for title in item.find_all(name='title'):
          jobs.append(title)
    return(jobs)
extract_job_title_from_result(soup)

```
